```python
img2 = Image.open('test.jpg')
array2 = np.asarray(img2)
```

```python
f = open('image.h', 'w')
f.write('const float image[3][244][244] = {\n')
for c in range(3):
    f.write('    {\n')
    for x in range(224):
        f.write('        {')
        for y in range(224):
            f.write(str(array2[x][y][c]) + ', ')
        f.write('}, \n')
    f.write('    },\n')
f.write('}\n\n')

f.close()
```

```python
f = open('image.hex', 'w')

for x in range(224):
    for y in range(224):
        f.write('00')
        for c in range(3):
            f.write(str(hex(array2[x][y][c]))[2:].rjust(2, '0'))
        f.write('\n')

f.close()
```
